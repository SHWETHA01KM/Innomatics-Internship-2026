```python
#Program1-Employee Performance Bonus Eligibility

employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

#Find the highest performance score
highest_score = max(employees.values())

#Identify all employees who have the highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == highest_score]

print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {highest_score})")
```

```python
#Program2-Search Query Keyword Analysis

query = input("Enter a search query: ")
query = query.lower()

#Remove common punctuation
import string
query = query.translate(str.maketrans('', '', string.punctuation))

#Split sentence into words
words = query.split()

#Count frequency of each keyword
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

result = {word: count for word, count in frequency.items() if count > 1}

print(result)
```

```python
#Program3-Sensor Data Validation

# Sensor readings list (index = hour, value = reading)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

#Identify even readings and store as (hour_index, reading_value) pairs
valid_readings = []

for hour, value in enumerate(sensor_readings):
    if value % 2 == 0:   # Check if reading is even (valid)
        valid_readings.append((hour, value))


print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Program4- Email Domain Usage Analysis
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

#Extract domains
domains = [email.split("@")[1] for email in emails]

#Count domain occurrences using dictionary comprehension
domain_count = {domain: domains.count(domain) for domain in set(domains)}

total_users = len(emails)

print("Email Domain Usage Percentage:")

for domain, count in domain_count.items():
    percentage = (count / total_users) * 100
    print(f"{domain}: {int(percentage)}%")
```

```python
#Program5-Sales Spike Detection

#Given daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

#Calculate average daily sales
average_sales = sum(sales) / len(sales)

#Determine spike threshold (30% above average)
threshold = average_sales + (0.30 * average_sales)

print("Sales Spike Days:")

for day, value in enumerate(sales, start=1):  #start=1 for day numbering
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
#Program6-Duplicate User ID Detection

# Given list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

#Count occurrences of each user ID
id_count = {}

for user in user_ids:
    if user in id_count:
        id_count[user] += 1
    else:
        id_count[user] = 1

#Display only duplicate IDs (count > 1)
print("Duplicate User IDs:")

for user, count in id_count.items():
    if count > 1:
        print(f"{user} → {count} times")
```
