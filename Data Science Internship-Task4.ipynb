```python
#Problem1: Mobile Recharge Validation System
def recharge(amount):
  vaild_plan=[199,299,399,599]

  if amount<50:
    print("Recharge amount must be more than 50")
    return False
  if amount not in vaild_plan:
    print("Invaild plan")
    return False
  return True

while True:
  amount=int(input("Enter the Recharge Amount:"))

  if recharge(amount):
    print("Recharge is successful")
    break
  else:
    print("Try Again")
```

```python
#Problem2: Inventory Reorder Alert System
def reorder_alert(product):
  for item in product:
    stock = product[item]

    if stock < 15:
      print("Reorder Alert")
    else:
      print("Stock is ok")
  return

product={'kurthi':56,'frock':34,'jeans':10,'shirt':14}

reorder_alert(product)
```

```python
#Program3: Student Result Processing System
def calculate_result(marks):

    total = 0

    # Calculate total marks using loop
    for mark in marks:
        total += mark

    average = total / len(marks)

    # Check pass/fail condition
    if average >= 50:
        return "Pass"
    else:
        return "Fail"


num_subjects = int(input("Enter number of subjects: "))
marks_list = []

for i in range(num_subjects):
    mark = float(input(f"Enter mark {i+1}: "))
    marks_list.append(mark)

result = calculate_result(marks_list)

print("Final Result:", result)
```

```python
#Program4: Cab Fare Estimator

def calculate_fare(distance, is_peak):

    base_fare = 50
    rate_per_km = 12

    fare = base_fare + (rate_per_km * distance)

    # Apply peak hour surcharge
    if is_peak.lower() == "yes":
        fare += fare * 0.25

    return fare

while True:

    distance = float(input("Enter distance travelled (km): "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak_hour)

    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")

    if retry.lower() != "yes":
        print("Thank you for using Cab Fare Estimator!")
        break
```

```python
#Program5: Attendance Eligibility Checker
def check_attendance(attendance_list):
    present_days = 0

    # Count number of present days
    for day in attendance_list:
        if day.upper() == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

total_days = int(input("Enter total working days: "))
attendance_records = []

for i in range(total_days):
    status = input(f"Enter attendance for day {i+1} (P/A): ")
    attendance_records.append(status)

eligibility = check_attendance(attendance_records)

print("Attendance Status:", eligibility)
```

```python
#Program6: Password Strength Checker

def check_password_strength(password):
    special_characters = "@#$"
    has_digit = False
    has_special = False

    # Check minimum length
    if len(password) < 8:
        return "Weak Password"

    # Check each character
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_characters:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

user_password = input("Enter your password: ")

strength = check_password_strength(user_password)

print("Password Strength:", strength)
```
