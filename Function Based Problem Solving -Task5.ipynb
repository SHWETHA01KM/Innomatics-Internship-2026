```python
#Problem1:Social Media – Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0

    #Calculate total likes using loop
    for like in likes_list:
        total_likes += like

    #Check post status
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)


likes = list(map(int, input("Enter likes separated by space: ").split()))

analyze_engagement(likes)
```

```python
#Problem2: Healthcare – Medicine Stock Alert System

#function to check medicine stock
def check_stock(stock):

    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


stock = int(input("Enter medicine stock: "))

check_stock(stock)
```

```python
#Problem3: Agriculture – Rainfall Adequacy Checker

# Function to check rainfall adequacy
def rainfall_checker(average):

    required_level = 70   # Fixed required rainfall level

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

average_rainfall = int(input("Enter average rainfall: "))

rainfall_checker(average_rainfall)
```

```python
# Problem 4: Social Media – Duplicate Account Detection

# Function to detect duplicate usernames
def check_duplicates(usernames):

    # Using set to remove duplicates
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)

users = ["Ram", "rahul", "ananya", "Ram"]

check_duplicates(users)
```

```python
#Problem 5: Healthcare – Appointment Eligibility Checker

# Function to check appointment eligibility
def check_eligibility(age):

    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

age = int(input("Enter patient age: "))

check_eligibility(age)
```

```python
# Problem 6: Agriculture – Premium Crop Price Filter

# Function to filter premium crops
def filter_premium_crops(price_list):
    premium = []

    # Filtering logic using loop
    for price in price_list:
        if price > 2000:
            premium.append(price)

    print("Premium Crops:", premium)

# Predefined crop prices
prices = [1500, 2500, 1800, 3200, 1950]

filter_premium_crops(prices)
```

```python
#Problem 7: System Monitoring – Application Health Checker

# Function to check system health
def check_health(error_count):

    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)

errors = int(input("Enter error count: "))

check_health(errors)
```

```python
#Problem 8: Banking – Daily Transaction Limit Checker

# Function to check transaction limit
def check_transaction(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


amount = int(input("Enter transaction amount: "))

check_transaction(amount)
```

```python
#Problem 10: Smart Electricity Bill Analyzer

# Function to calculate electricity bill
def calculate_bill(units):
    bill = 0

    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage classification
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


units = int(input("Enter electricity units consumed: "))

bill, usage_status = calculate_bill(units)

print("Total Bill:", bill)
print("Usage Status:", usage_status)
```
