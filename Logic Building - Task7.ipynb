```python
#Problem 1: Smart Light Controller

class SmartLight:
    #Class to manage smart light operations.

    def __init__(self, name):
        # Initialize light name and default status
        self.name = name
        self.status = "OFF"

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


def main():
    light_name = input("Enter Light Name: ")
    action = input("Enter Action (ON/OFF): ").upper()

    light = SmartLight(light_name)

    if action == "ON":
        light.turn_on()
    elif action == "OFF":
        light.turn_off()
    else:
        print("Invalid action! Defaulting to OFF.")

    light.display_status()


if __name__ == "__main__":
    main()
```

```python
#Problem 2: Employee ID Card System
class Employee:
    #Class to store employee information.

    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("\nEmployee ID Card")
        print(f"Name       : {self.name}")
        print(f"ID         : {self.emp_id}")
        print(f"Department : {self.department}")


def main():
    name = input("Enter Employee Name: ")
    emp_id = input("Enter Employee ID: ")
    department = input("Enter Department: ")

    employee = Employee(name, emp_id, department)
    employee.display_id_card()


if __name__ == "__main__":
    main()
```

```python
#Problem 3: Mobile Contact Record

class Contact:
    #Class to manage contact details.

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("\nContact Saved")
        print(f"Name  : {self.name}")
        print(f"Phone : {self.phone}")


def main():
    name = input("Enter Contact Name: ")
    phone = input("Enter Phone Number: ")

    contact = Contact(name, phone)
    contact.display_contact()


if __name__ == "__main__":
    main()
```

```python
#Problem 4: Product Price Tag Generator

class Product:
    #Class to handle product price details.

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("\nPrice Tag")
        print(f"Product : {self.name}")
        print(f"Price   : ₹{self.price}")


def main():
    name = input("Enter Product Name: ")
    price = input("Enter Price: ")

    product = Product(name, price)
    product.display_price_tag()


if __name__ == "__main__":
    main()
```

```python
#Problem 5: Movie Rating Display System

class Movie:
    #Class to manage movie rating details.

    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print("\nMovie Details")
        print(f"Movie  : {self.name}")
        print(f"Rating : {self.rating} / 5")


def main():
    name = input("Enter Movie Name: ")
    rating = input("Enter Rating: ")

    movie = Movie(name, rating)
    movie.display_movie()


if __name__ == "__main__":
    main()
```

```python
#Problem 6: Delivery Address Manager

class Delivery:
    #Class to manage delivery details.

    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display_details(self):
        print("\nDelivery Details")
        print(f"Customer : {self.customer}")
        print(f"Address  : {self.address}")


def main():
    customer = input("Enter Customer Name: ")
    address = input("Enter Address: ")

    delivery = Delivery(customer, address)
    delivery.display_details()


if __name__ == "__main__":
    main()
```
