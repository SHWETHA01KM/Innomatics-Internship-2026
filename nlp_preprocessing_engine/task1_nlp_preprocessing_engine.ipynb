# Task 1: Conceptual Understanding

## 1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are treated as two different words because one is capitalized and the other is not. This can increase the number of unique words unnecessarily. To avoid this, we usually convert all text to lowercase so that both are treated the same.



## 2. What happens if stopwords are not removed?

If stopwords are not removed, the data becomes noisy. Words like "is", "the", and "and" appear very frequently but do not add much meaning. Because of this, the model may focus more on these common words instead of important ones, which can reduce performance.



## 3. When can removing stopwords be harmful?

Removing stopwords is not always good.

- In sentiment analysis, words like "not" are very important. For example, "I am not happy" will change meaning if "not" is removed.  
- In question answering or chatbots, removing words like "what", "is" can make the sentence incomplete or confusing.



## 4. Difference between Stemming and Lemmatization

Stemming and lemmatization are both used to reduce words to their base form.

- Stemming simply cuts the word based on rules, so sometimes the output is not a proper word. For example, "running" becomes "runn".  
- Lemmatization uses a dictionary and returns the correct base word. For example, "running" becomes "run".  

So, lemmatization is more accurate, but stemming is faster.

# Task 2: Build Advanced Preprocessing Function

```python
import re

def preprocess_text(text):
    try:
        # Handle empty or invalid input
        if not text or not isinstance(text, str):
            return [], ""

        # Convert to lowercase
        text = text.lower()

        # Remove URLs
        text = re.sub(r"http\S+|www\S+", "", text)

        # Remove email patterns
        text = re.sub(r"\S+@\S+", "", text)

        # Remove numbers
        text = re.sub(r"\d+", "", text)

        # Handle repeated characters
        text = re.sub(r'(.)\1{2,}', r'\1\1', text)
        text = re.sub(r'oo+', 'o', text)

        # Remove special characters and emojis
        text = re.sub(r"[^a-z\s]", "", text)

        # Remove extra spaces
        text = re.sub(r"\s+", " ", text).strip()

        # Tokenization
        tokens = text.split()

        # Remove short words (<=2) except 'no' and 'not'
        tokens = [word for word in tokens if len(word) > 2 or word in ['no', 'not']]

        # Join back to sentence
        clean_sentence = " ".join(tokens)

        return tokens, clean_sentence

    except Exception as e:
        print("Error:", e)
        return [], ""
```

### Sample Test

```python
test_sentence = "I absolutely looooved this product"

tokens, clean_sentence = preprocess_text(test_sentence)

print("Original Text:", test_sentence)
print("Cleaned Tokens:", tokens)
print("Cleaned Sentence:", clean_sentence)
```

# Task 3: Stress Testing


```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for i, sentence in enumerate(test_sentences, 1):
    tokens, clean_sentence = preprocess_text(sentence)

    print(f"Sentence {i}")
    print("Original Text:", sentence)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("\n")
```

# Task 4: Token Analytics

```python
import numpy as np

for i, sentence in enumerate(test_sentences, 1):
    tokens, _ = preprocess_text(sentence)

    if len(tokens) == 0:
        print(f"Sentence {i}: No tokens available")
        print("\n")
        continue

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens])

    print(f"Sentence {i}")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Average Token Length:", round(avg_length, 2))
    print("\n")
```

## Which sentence had the most noise?

The sentence Win \$\$\$ now!!! Limited offer!!! had the most noise because it contains a lot of special symbols like "$$$" and repeated exclamation marks. These do not add meaningful information and were removed during preprocessing.

## Which sentence retained the most meaningful tokens after cleaning?

The sentence "I am not happy with this" retained the most meaningful tokens. Important words like "not" and "happy" were preserved, which clearly express the sentiment even after cleaning.

# Task 5: Frequency Analysis

```python
from collections import Counter

# Combine all tokens
all_tokens = []
for sentence in test_sentences:
    tokens, _ = preprocess_text(sentence)
    all_tokens.extend(tokens)

# Count frequencies
counter = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

# Task 6: Full NLP Pipeline

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean_sentence = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

###Test the Pipeline

```python
result = full_pipeline(test_sentences)

print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
print(result["clean_sentences"])
```

# Task 7: Error Handling

```python
edge_cases = [
    "",
    "😍😍😍",
    "1234567890"
]

for case in edge_cases:
    tokens, clean_sentence = preprocess_text(case)

    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned Sentence:", clean_sentence)
    print("\n")
```
