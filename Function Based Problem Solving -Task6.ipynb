```python
#Problem 1: Smart Parking Lot Management System
def smart_parking_system(capacity, vehicle_logs):
    current_vehicles = 0

    for log in vehicle_logs:
        log = log.strip().upper()

        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1

    print("Currently Parked Vehicles:", current_vehicles)

    if current_vehicles > capacity:
        print("Parking Status: Exceeded Capacity!")
    else:
        print("Parking Status: Available")

capacity = int(input("Parking Capacity: "))
logs_input = input("Vehicle Logs (comma separated like IN,IN,OUT): ")

vehicle_logs = logs_input.split()

smart_parking_system(capacity, vehicle_logs)
```

```python
# Problem 2: Online Food Delivery Time Estimator

def estimate_delivery_time(distance, traffic, weather):

    # Base time (5 minutes per km)
    eta = distance * 5

    # Traffic delay
    traffic_delay = 0
    traffic = traffic.strip().lower()

    if traffic == "high":
        traffic_delay = 15
    elif traffic == "medium":
        traffic_delay = 10
    elif traffic == "low":
        traffic_delay = 5

    # Weather delay
    weather_delay = 0
    weather = weather.strip().lower()

    if weather == "rainy":
        weather_delay = 10
    elif weather == "stormy":
        weather_delay = 20

    # Apply only maximum delay
    eta += max(traffic_delay, weather_delay)

    print("Estimated Delivery Time:", eta, "minutes")


distance = int(input("Distance (km): "))
traffic = input("Traffic Level: ")
weather = input("Weather: ")

estimate_delivery_time(distance, traffic, weather)
```

```python
# Problem 3: Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats, booked_count):
    """
    Calculates occupancy percentage and determines show status.
    """

    # Prevent invalid input
    if total_seats <= 0:
        print("Invalid total seats.")
        return

    if booked_count > total_seats:
        print("Error: Booked seats cannot exceed total seats.")
        return

    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100

    print(f"Occupancy: {round(occupancy)}%")
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
total_seats = int(input("Total Seats: "))
booked_count = int(input("Number of Booked Seats: "))

seat_occupancy(total_seats, booked_count)
```

```python
# Problem 4: Cloud Server Load Classification System

def classify_server_load(cpu_readings):
    #Classifies server load based on average CPU usage.

    # Prevent empty list error
    if len(cpu_readings) == 0:
        print("No CPU readings provided.")
        return

    total = 0

    # Calculate total CPU usage
    for value in cpu_readings:
        total += value

    # Calculate average
    average = total / len(cpu_readings)

    print(f"Average CPU Load: {round(average)}%")

    # Classification logic
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


cpu_input = input("CPU Readings (%): ")

cpu_readings = [int(value.strip()) for value in cpu_input.split()]

classify_server_load(cpu_readings)
```

```python
# Problem 5: Smart Classroom Resource Usage Monitor
def classroom_resource_monitor(resource_usage):

    #Identifies overused classroom resources.
    #Overuse condition: Usage more than 8 hours.

    overused = []
    # Check each resource usage
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resource_usage)
```

```python
# Problem 6: Online Event Registration Capacity Controller

def event_registration_controller(capacity, registrations):
   # Controls event registrations.
   # Prevents overbooking and activates waitlist mode.


    # If registrations exceed capacity
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

capacity = int(input("Event Capacity: "))
registrations = int(input("Registrations: "))

event_registration_controller(capacity, registrations)
```
